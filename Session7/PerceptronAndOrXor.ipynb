```
#HW question 1

#Code perceptron from scratch for AND and OR gate.
import numpy as np

def loss(expected,actual):
  return expected-actual

def activation(threshold,val):
  if(actual_op>threshold):
    return 1
  else:
    return 0

#And gate
And=[[0,0],[0,1]]
Or=[[0,1],[1,1]]


weight= [0.3,-0.1]
threshold=0.2
error=10000
rate=0.1

print("AND\n")
epoch=0
while error>0 and epoch<21:
  epoch+=1
  error=0
  for i in range(2):
    for j in range(2):
      actual_op=weight[0]*i + weight[1]*j
      actual_op=activation(threshold,actual_op)
      err=loss(And[i][j],actual_op)
      err=And[i][j]-actual_op
      # print(err)
      # print(rate*(err)*)
      weight[0]=weight[0]+rate*(err)*i
      weight[1]=weight[1]+rate*(err)*j
      error+=abs(err)
      print('epoch',epoch,'input',i,j,"threshold",threshold,'expected output',And[i][j],'Actual output',actual_op,'Weights',weight[0],weight[1],'Error',err)
print("Final ouptut",weight,"with threshold",threshold,'\n\n\n')

print("Or")
error=1000
epoch=0
weight=[0.3,-0.1]
while error>0 and epoch<21:
  epoch+=1
  error=0
  for i in range(2):
    for j in range(2):
      actual_op=weight[0]*i + weight[1]*j
      actual_op=activation(threshold,actual_op)

      err=loss(Or[i][j],actual_op)
      # print(err)
      # print(rate*(err)*)
      weight[0]=weight[0]+rate*(err)*i
      weight[1]=weight[1]+rate*(err)*j
      error+=abs(err)
      print('epoch',epoch,'input',i,j,"threshold",threshold,'expected output',Or[i][j],'Actual output',actual_op,'Weights',weight[0],weight[1],'Error',err)
print("Final ouptut",weight,"with threshold",threshold,'\n\n\n')


  
print("Xor early stop")

Xor=[[0,1],[1,0]]

weight=[0.3,-0.1]
epoch=0
error=1000
sameval=0
while error>0 and sameval<=5:
  epoch+=1
  error=0
  preverror=error
  for i in range(2):
    for j in range(2):
      actual_op=weight[0]*i + weight[1]*j
      actual_op=activation(threshold,actual_op)
      err=loss(Xor[i][j],actual_op)
      # print(err)
      # print(rate*(err)*)
      weight[0]=weight[0]+rate*(err)*i
      weight[1]=weight[1]+rate*(err)*j
      error+=abs(err)
      print('epoch',epoch,'input',i,j,"threshold",threshold,'expected output',Xor[i][j],'Actual output',actual_op,'Weights',weight[0],weight[1],'Error',err)
  if(preverror<=error):
    sameval+=1
print("After patience value exceeded 5,loop has stopped early")

```
