We know from the mid school mathematics that the distance from origin to the line $ax + by + c = 0$ is $\frac{|c|}{\sqrt{a^2 + b^2}}$.

Equivalently, distance from the origin to the line $w^TX + c = 0$ is given by $\frac{|-c|}{||w||}$

**Question:**
Can you write code to find distance of the line $x = 50$ and $y = 25$ from the origin?

Can you derive the formula to compute distance between two parallel lines from the formulas provided in the previous section?

```
# Write your solution here
def distance_from_origin(a, b, c):
    return abs(c)/(np.sqrt(np.square(a)+np.square(b)))

print(distance_from_origin(1,0,-50))
print(distance_from_origin(0,1,-25))

#Distance between parallel lines
def pardist(a,b,c1,c2):
  return abs(distance_from_origin(a,b,c1)-distance_from_origin(a,b,c2))

print(pardist(1,0,40,50))
```

**tanh function**
\begin{equation}
tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
\end{equation}

**Gaussian Function**:
\begin{equation}
f(x) = e^{-x^2}
\end{equation}

**Rectified Linear Unit(ReLU)**:
\begin{equation}
f(x) = max(0, x)
\end{equation}

```
#sigmoid
X=np.linspace(-10,10,100)
Y=1/(1+np.exp(-X))
plt.plot(X,Y)
```

```
#tanh
X=np.linspace(-10,10,100)
Y=(np.exp(X)-np.exp(-X))/(np.exp(X)+np.exp(-X))
plt.plot(X,Y)
```

```
# Plotting a 3D surface
from mpl_toolkits.mplot3d import Axes3D # Importing this is necessary when you plot in 3D
a = 1
b = 1
c = 3
d = 10
##############################
xx = np.linspace(-10, 10, 20)
yy = np.linspace(-10, 10, 20)
X, Y = np.meshgrid(xx, yy)
zz = (-a*X - b*Y - d) * 1.0 / c
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(X, Y, zz)
```
